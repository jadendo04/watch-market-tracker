```
import pandas as pd
from sqlalchemy import create_engine

engine = create_engine("postgresql://dvt@localhost:5432/watch_tracker")
df = pd.read_sql("SELECT * FROM watches LIMIT 10000", engine)
print(df.shape)
df.head()
```

```
missing = df.isnull().sum().sort_values(ascending=False)
print(missing)
```

```
import seaborn as sns
import matplotlib.pyplot as plt

df_filtered = df[df['price'] < 100000]
sns.histplot(df_filtered['price'], bins=50)
plt.title('Price Distribution (under $100k)')
plt.xlabel('Price ($)')
plt.show()
```

```
top_brands = df['brand'].value_counts().head(10)
sns.barplot(x=top_brands.values, y=top_brands.index)
plt.title('Top 10 Brands by Listings')
plt.xlabel('Number of Listings')
plt.show()
```

```
avg_cond = df.groupby('cond')['price'].mean().sort_values(ascending=False)
sns.barplot(x=avg_cond.values, y=avg_cond.index)
plt.title('Average Price by Condition')
plt.xlabel('Avg Price ($)')
plt.show()
```
